```python
from selenium import webdriver as driver
from selenium.webdriver.common.keys import Keys
from time import sleep
import pandas as pd
from tkinter import *

from selenium.webdriver.common.by import By

#VAR
color_bg = '#282A36'
font = 'arial 11'
color_fg = 'white'
color_bt = '#4F729A'
endereco = []
    
def create_dataFrame(dados):

    for p, dado in enumerate(dados):
        endereco.append(dado.text)

    txt_logradouro.config(text=f'Logradou: {endereco[0]}')
    txt_bairro.config(text=f'Bairro: {endereco[0]}')
    txt_uf.config(text=f'UF: {endereco[0]}')
    txt_cep.config(text=f'CEP: {endereco[0]}')

    end_ = [endereco.copy()]

    #criando dataframe com os dados
    df = pd.DataFrame(end_, columns=['logradouro','bairro','UF','cep'])  
    
    #df.to_csv('correios.csv', index=None, encoding='utf-8')
    
    print(df)

    return print('df salvo!')

def search_cep():
    cep = input_pesquia.get()

    options = driver.ChromeOptions()
    options.add_argument('--headless=new')

    browser = driver.Chrome(options=options)
    browser.get('https://buscacepinter.correios.com.br/app/endereco/index.php')

    browser.find_element(By.NAME, 'endereco').send_keys(cep)

    sleep(1)

    #Clicando em pesquisar
    browser.find_element(By.NAME, 'btn_pesquisar').click()

    sleep(2)

    dados =  browser.find_elements(By.XPATH,'//td')

    create_dataFrame(dados)

root = Tk()
root.config(background=color_bg)
root.title('Buscar Cep')
root.geometry('300x300+500+200')

txt_cep = Label(root,
              text='Cep',
              bg=color_bg,
              fg=color_fg,
              font=font).grid(row=0,column=0)

txt_logradouro = Label(root,
              text=f'Logradouro: ',
              font=font,
              bg=color_bg,
              fg=color_fg)
txt_logradouro.grid(row=1,column=0,columnspan=2)

txt_bairro = Label(root,
              text=f'Bairro:',
              font=font,
              bg=color_bg,
              fg=color_fg)
txt_bairro.grid(row=2,column=0,columnspan=2)

txt_uf = Label(root,
              text=f'UF:',
              font=font,
              bg=color_bg,
              fg=color_fg)
txt_uf.grid(row=3,column=0,columnspan=2)

txt_cep = Label(root,
              text=f'CEP:',
              font=font,
              bg=color_bg,
              fg=color_fg)
txt_cep.grid(row=4,column=0,columnspan=2)

button_pesquisar = Button(root,
                          text='pesquisar',
                          bg=color_bt,
                          fg=color_fg,
                          font=font,
                          width=10,
                          height=1,
                          command=search_cep).grid(row=5,column=0,columnspan=2)

input_pesquia = Entry(root)
input_pesquia.grid(row=0,column=1)

root.mainloop()
```
